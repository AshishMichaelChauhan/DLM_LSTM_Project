## 3. ‚öôÔ∏è **Analysis and Approach**

### 3.1 **Data Collection and Preprocessing**
- **Data Source:** Yahoo Finance (via `yfinance` package).  
- **Stock Tickers:**  
    - NIFTY 50: `^NSEI` (National Stock Exchange Index)  
    - Nestle India Ltd.: `NESTLEIND.NS`  
- **Date Range:**  
    - **2005-01-31** to **2025-01-31** (20 years of historical data)  
- **Data Focus:**  
    - Only **closing prices** were extracted, as they provide a reliable summary of the stock‚Äôs performance.  
    - Data was stored in a Pandas DataFrame and normalized using **MinMaxScaler** to scale values between **0 and 1**, improving model efficiency.  

---

### 3.2 **Sequence Creation for LSTM Model**
- **Why Sequences?**  
    - Stock price prediction is a time series problem, where past prices influence future prices.  
    - LSTM models require sequences of past data to predict future values.  
- **Sequence Length:**  
    - A window size of **60 days** was chosen, meaning the model looks at the past 60 days‚Äô closing prices to predict the next day‚Äôs price.  
- **Data Preparation:**  
    - Sequences were created using a **sliding window approach**.  
    - Each sequence contains:  
        - **Input (X):** 60 consecutive closing prices.  
        - **Target (y):** The next day's closing price.  

---

### 3.3 **Splitting Data for Training and Testing**
- **Train-Test Split:**  
    - **80%** of the sequences were used for training.  
    - **20%** were used for testing.  
- **Reshape for LSTM Input:**  
    - The data was reshaped into the format required by LSTM:  
    ```scss
    (batch_size, time_steps, features)
    ```

## 4. üîß **Building and Configuring the LSTM Model**

### 4.1 **Why Use LSTM for Stock Prediction?**
- **LSTM** networks are a specialized type of **Recurrent Neural Network (RNN)** designed to remember **long-term dependencies**.  
- They are highly effective for **time series problems**, where historical data influences future outcomes.  

### 4.2 **Model Architecture**
- **Input Layer:**  
    - 60 time steps (past 60 days) with **1 feature** (closing price).  
- **LSTM Layers:**  
    - 3 LSTM layers with **50 units** each.  
    - **Dropout(0.2)** applied after each LSTM layer to prevent overfitting.  
- **Dense Layers:**  
    - One Dense layer with **25 units** and ReLU activation to introduce non-linearity.  
    - Final Dense layer with **1 unit** to predict the next day‚Äôs closing price.  

---

### 4.3 **Model Compilation**
- **Optimizer:**  
    - Adam (`learning rate = 0.001`) for optimal convergence.  
- **Loss Function:**  
    - Mean Squared Error (**MSE**) for regression tasks.  
- **Metrics:**  
    - Mean Absolute Error (**MAE**) to assess model performance.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
````

````python
# Download NIFTY 50 data (last 20 years)
nifty_data = yf.download('^NSEI', start='2005-01-31', end='2025-01-31')
````

````python
# Extract closing prices
data = nifty_data[['Close']]
````

````python
# Normalize data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)
````

````python
#show data
data.head()
````

````python
# Prepare training data
def create_sequences(data, time_step):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)
````

````python
time_step = 60  # Using past 60 days for prediction
X, y = create_sequences(data_scaled, time_step)
````

````python
# Split into training and testing sets
split_ratio = 0.8
split = int(len(X) * split_ratio)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
````

````python
# Reshape for LSTM input
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
````

````python
# Build LSTM Model
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(time_step, 1)),
    Dropout(0.2),
    LSTM(units=50, return_sequences=True),
    Dropout(0.2),
    LSTM(units=50),
    Dropout(0.2),
    Dense(units=25, activation='relu'),
    Dense(units=1)
])
````

````python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mae'])
````

````python
# Define accuracy metric (Mean Absolute Percentage Error - MAPE)
def mape(y_true, y_pred_nifty):
    y_true, y_pred_nifty = np.array(y_true), np.array(y_pred_nifty)
    nonzero_idx = y_true != 0  # Avoid division by zero
    return np.mean(np.abs((y_true[nonzero_idx] - y_pred_nifty[nonzero_idx]) / y_true[nonzero_idx])) * 100
````

````python
# Custom callback to print loss and accuracy after each epoch
class EpochCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        print(f"Epoch {epoch+1}: Loss = {logs['loss']:.4f}, Val Loss = {logs['val_loss']:.4f}, MAE = {logs['mae']:.4f}, Val MAE = {logs['val_mae']:.4f}")
````

````python
# Train the model and store training history
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test), callbacks=[EpochCallback()])
````

````python
# Predict on test data
y_pred_nifty = model.predict(X_test)
y_pred_nifty = scaler.inverse_transform(y_pred_nifty.reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
````

````python
# Calculate accuracy using MAPE
accuracy = 100 - mape(y_test_actual, y_pred_nifty)
print(f"Final Model Accuracy: {accuracy:.2f}%")
````

````python
# Plot results year-wise
plt.figure(figsize=(12,6))
plt.plot(nifty_data.index[split+time_step+1:], y_test_actual, label='Actual Price')
plt.plot(nifty_data.index[split+time_step+1:], y_pred_nifty, label='Predicted Price')
plt.xlabel('Year')
plt.ylabel('NIFTY 50 Price')
plt.xticks(rotation=45)
plt.legend()
plt.title('NIFTY 50 Price Prediction using LSTM (Year-wise)')
plt.show()
````

````python
# Download Nestle India Ltd. stock data (last 20 years)
nestle_data = yf.download('NESTLEIND.NS', start='2005-01-31', end='2025-01-31')

# Extract closing prices
data = nestle_data[['Close']]

data['Year'] = data.index.year  # Extract year from index

# Normalize data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data[['Close']])

# Prepare training data
def create_sequences(data, time_step):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 60  # Using past 60 days for prediction
X, y = create_sequences(data_scaled, time_step)

# Split into training and testing sets
split_ratio = 0.8
split = int(len(X) * split_ratio)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Reshape for LSTM input
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Build LSTM Model
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(time_step, 1)),
    Dropout(0.2),
    LSTM(units=50, return_sequences=True),
    Dropout(0.2),
    LSTM(units=50),
    Dropout(0.2),
    Dense(units=25, activation='relu'),
    Dense(units=1)
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mae'])

# Define accuracy metric (Mean Absolute Percentage Error - MAPE)
def mape(y_true, y_pred_nestle):
    y_true, y_pred_nestle = np.array(y_true), np.array(y_pred_nestle)
    nonzero_idx = y_true != 0  # Avoid division by zero
    return np.mean(np.abs((y_true[nonzero_idx] - y_pred_nestle[nonzero_idx]) / y_true[nonzero_idx])) * 100

# Custom callback to print loss and accuracy after each epoch
class EpochCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        print(f"Epoch {epoch+1}: Loss = {logs['loss']:.4f}, Val Loss = {logs['val_loss']:.4f}, MAE = {logs['mae']:.4f}, Val MAE = {logs['val_mae']:.4f}")

# Train the model and store training history
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test), callbacks=[EpochCallback()])

# Predict on test data
y_pred_nestle = model.predict(X_test)
y_pred_nestle = scaler.inverse_transform(y_pred_nestle.reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate accuracy using MAPE
accuracy = 100 - mape(y_test_actual, y_pred_nestle)
print(f"Final Model Accuracy: {accuracy:.2f}%")

# Plot results year-wise
plt.figure(figsize=(12,6))
plt.plot(nestle_data.index[split+time_step+1:], y_test_actual, label='Actual Price')
plt.plot(nestle_data.index[split+time_step+1:], y_pred_nestle, label='Predicted Price')
plt.xlabel('Year')
plt.ylabel('Nestle India Stock Price')
plt.xticks(rotation=45)
plt.legend()
plt.title('Nestle India Stock Price Prediction using LSTM (Year-wise)')
plt.show()
````
